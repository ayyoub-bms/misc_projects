# Importance sampling and deep out of the money option pricing (a toy example)

In this notebook we show how one can use the probability measure change to price deep out of the money options using monte-carlo.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from research.pricing.blackscholes import bsm, dplus
```

Consider a call option with strike prices ranging from $K=40$ to $K=100$ on an underlying asset with current price $S_t = 40$. And assume the following values for the remaining parameters.

```python
T = .8
K = np.array([40, 50, 60, 70, 80, 90, 100])
St = 40
σ = .15
r = .002
t = 0
```

## The BSM price

We start by computing the Black Scholes price using the closed form solution

```python
bs_prices = bsm(t, St, T, K, r, σ)['C']
```

We get a price per option:

```python
bs_prices
```

## Using importance sampling


Since the option is deeper out of the money, we would need a huge number of simulations to obtain points in the tails of the distribution.  Shifting the mean of the distribution will allow us to obtain points with 50% probability. To do so, we shift the mean of the standard normal so that it is exactly where we need it: at $d_+$ (or $d_-$ thanks to the symmetry of a standard gaussian). 

Using a change of measure by affine transformation for Gaussians, (or Girsanov theorem), we get the appropriate density of measure change:

$$ L = \frac{d\mathbb{Q}}{d\mathbb{P}} = e^{d_1Z - \frac{d_1^2}{2}} $$

where $ d_1 = \frac{\log{\frac{K}{Se^{r(T-t)}}}}{\sigma \sqrt{T-t}} + \frac{\sigma \sqrt{T-t}}{2}$ and $Z \sim \mathcal{N}(0,1)$

```python
# no simulations
no_sims = [1000*i for i in range(1, 1000, 10)]

def price(Z):
    return St * np.exp((r - σ**2/2)*(T-t) + σ * np.sqrt(T-t) * Z)

def payoff(S):
    return np.exp(-r*(T-t))*np.maximum(S - K, 0)
```

```python
μ = dplus(t, St, T, K, r, σ, 0)
```

```python
crud_prices = []
imps_prices = []
imps_vars = []
crud_vars = []
for m in no_sims:
    val = payoff(price(np.random.randn(m, 1)))
    crud_prices.append(val.mean(axis=0))
    crud_vars.append(val.std(axis=0)**2)
    Z = np.random.randn(m, 1)
    L = np.exp(μ * Z - μ**2 / 2)
    val =(payoff(price(-μ + Z))*L)
    imps_prices.append(val.mean(axis=0))
    imps_vars.append(val.std(axis=0)**2)
```

```python
imps_prices = np.asarray(imps_prices)
crud_prices = np.asarray(crud_prices)
```

```python
for i in range(7):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[10, 4])
    fig.suptitle(f'Call price Monte Carlo simulations K = {K[i]}')

    ax1.axhline(bs_prices[i], lw=1.4, color= 'red', label='Analytic');
    ax1.plot(no_sims, imps_prices[:, i], color='darkgreen', alpha=.7, label='IS MC');
    ax1.legend(loc='upper right',  ncol=5);

    ax2.axhline(bs_prices[i], lw=1.4, color= 'red', label='Analytic');
    ax2.plot(no_sims, crud_prices[:, i], color='black', alpha=.7, label='Crude MC');
    ax2.legend(loc='upper right',  ncol=5);
```

```python
import pandas as pd
```

```python
pd.set_option('display.float_format', '{:.3E}'.format)
```

```python
ts = 90

summary = pd.DataFrame(columns=['True value', 'Crude MC value', 'Importance MC value'], index=K)
```

```python
summary['True value'] = bs_prices
summary['Crude MC value'] = crud_prices[ts]
summary['Importance MC value'] = imps_prices[ts]
summary['Crude MC Conf Int Lower bound'] = crud_prices[ts] - 1.96 * np.sqrt(crud_vars[ts] / no_sims[ts])
summary['Crude MC Conf Int Upper bound'] = crud_prices[ts] + 1.96 * np.sqrt(crud_vars[ts] / no_sims[ts])
summary['Importance MC Conf Int Lower bound'] = imps_prices[ts] - 1.96 * np.sqrt(imps_vars[ts] / no_sims[ts])
summary['Importance MC Conf Int Upper bound'] = imps_prices[ts] + 1.96 * np.sqrt(imps_vars[ts] / no_sims[ts])
```

```python
summary
```

We clearly see that the crude monte carlo method either sets the price to zero or way above its correct price when the option goes deeper out of the money, this is due to the fact that even by sampling millions of realizations, only few payoffs will be positive.
